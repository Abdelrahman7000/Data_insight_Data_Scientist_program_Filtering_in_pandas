```python
import pandas as pd
```

```python
#importing our data and make the data type of Dt_Customer column to be of type date 
df=pd.read_csv('marketing_campaign.csv',sep='\t', 
                   index_col='ID', 
                   parse_dates=['Dt_Customer'])
```

```python
df.head()
```

```python
df.Education.value_counts()
```

```python
# getting only the year of birth that equals to 1990
df[df.Year_Birth.eq(1990)]
```

```python
years=[1990,1985,1966,1952]
df[df.Year_Birth.isin(years)]
```

```python
# Filter based on one condition
df[df['Education']=='Master']
```

```python
# Filter based on multiple conditions using or operator
df[(df['Education']=='Master')|(df['Education']=='PhD')]
```

```python
# Using and operator, filter the records based on the Education and the Marital_Status columns
df[(df['Education']=='Master')&(df['Marital_Status']=='Married')]
```

```python
# get all the values except for the Master value of the Education column
df[~(df['Education']=='Master')]
```

```python
df.Marital_Status.value_counts()
```

```python
# filter using the query method
df.query('Education =="Master"& Marital_Status=="Single"')
```

```python
df.loc[(df.Education=='Graduation')&(df.Marital_Status=='Single'),'Year_Birth':'Income']
```

```python
df.iloc[:7,0:4]
```

```python
# get the values of the Education column that only begin with the letter "M"
df[df['Education'].str[0]=='M']
```
